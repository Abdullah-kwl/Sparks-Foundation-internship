# Modeling

```python
X=df.Hours.values.reshape(-1,1)
Y=df.Scores.values.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

#### Linear Regression

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
lr_pred=lr.predict(x_test)

mse=mean_squared_error(y_test,lr_pred)
mae=mean_absolute_error(y_test,lr_pred)
score=r2_score(y_test,lr_pred)

print('Mean Squared Error =',mse)
print('Mean Absolute Error =',mae)
print('R2 Score =',score)
```

#### Lasso

```python
la=Lasso()
la.fit(x_train,y_train)
la_pred=la.predict(x_test)

mse=mean_squared_error(y_test,la_pred)
mae=mean_absolute_error(y_test,la_pred)
score=r2_score(y_test,la_pred)

print('Mean Squared Error =',mse)
print('Mean Absolute Error =',mae)
print('R2 Score =',score)
```

#### Ridge

```python
rg=Ridge()
rg.fit(x_train,y_train)
rg_pred=rg.predict(x_test)

mse=mean_squared_error(y_test,rg_pred)
mae=mean_absolute_error(y_test,rg_pred)
score=r2_score(y_test,rg_pred)

print('Mean Squared Error =',mse)
print('Mean Absolute Error =',mae)
print('R2 Score =',score)
```

### Conclusion:

* Linear Regression Giving the Best R2-Score

#### Plotting the Linear Regression line

```python
# Plotting the regression line
line = lr.coef_*X+lr.intercept_

# Plotting for the test data
plt.scatter(X, Y)
plt.plot(X, line,'r');
plt.title('Regression line of LinearRegression')
plt.show()
```
